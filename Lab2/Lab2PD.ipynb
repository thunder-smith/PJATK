```python
import numpy as np
import pandas as pd
data = pd.read_csv('drive/MyDrive/president_heights.csv')
heights = np.array(data['height(cm)'])

print("Mean height:        ", np.mean(heights))
print("Standard deviation: ", np.std(heights))
print("Maximum height:     ", np.max(heights))
print("Minimum height:     ", np.min(heights))
print("25th percentile:    ", np.percentile(heights, 25))
print("75th percentile:    ", np.percentile(heights, 75))
print("Median:             ", np.median(heights))
```

```python
rainfall = pd.read_csv('drive/MyDrive/Seattle2014.csv')['PRCP'].values
inches = rainfall / 254.0  # 1/10mm -> inches
inches.shape

rainy = np.count_nonzero(inches)
print("Rainy days:     ", rainy)
print("Non rainy days: ", inches.size - rainy)
print("Days above 0.5 inch of rain: ", np.count_nonzero(inches > 0.5))
print("Days below 0.2 inch of rain: ", np.count_nonzero(inches < 0.2))
```

```python
masked_rainy = np.ma.array(inches, mask=(inches <= 0))
print("Median precip on rainy days in 2014 (inches):   ", np.ma.median(masked_rainy)) 

masked_summer = np.ma.array(inches, mask = False)
m1 = (inches == 0)
m2 = np.ma.make_mask_none(m1.shape)
m2.fill(True)
m2[172:262] = False
masked_summer.mask = np.ma.mask_or(m1,m2)
print("Median precip on summer days in 2014 (inches):  ", np.ma.median(masked_summer))
print("Maximum precip on summer days in 2014 (inches): ", np.ma.max(masked_summer)) 

masked_no_summer = np.ma.array(masked_rainy, copy=True)
masked_no_summer.mask[172:262] = True
print("Median precip on non-summer rainy days (inches):", np.ma.median(masked_no_summer)) 
```
